## Tarefa 1- Métodos de strings

<div align='justify'>
Nesta primeira semana, vamos abordar alguns métodos que podem ser utilizados com strings, métodos que são bastante uteis quando é ncessário ao longo do codigo trabalhar com variaveis em formato string. Com o intuito de apresentar 3 métodos de strings, procedemos à criação de uma função que calcule o conteudo GC, em percentagem, numa cadeia de DNA.

A função gc_content recebe como argumento uma sequência de nucleoticos em formato string e devolve um float que corresponde ao conteúdo de guaninas e citosinas, em percentagem, da sequência. Nesta função utilizaram-se métodos de strings de modo a trata a sequência fornecida com o intuito de padronizar a cadeia e permitir a contagem de "C" e "G". O método upper() garante que todos os caracteres fornecidos estão em letra maiúscula, enquanto o método strip () remove possíveis espaços em branco que possam ter sido inseridos nas extremidades da cadeia, o que levaria a erros no cálculo do conteudo CG. Este método apresenta um parâmetro opcional onde se pode forncer, em formato de string, os caracteres que se pretende eliminar, sendo que, por padrão, o método remove espaços em branco. Por fim, aplicou-se o método replace() de modo a garantir que a cadeia não possui espaços no seu interior. O método replace() apresenta 3 parâmetros, sendo 2 deles obrigatórios. O primeiro parâmetro corresponde ao caracter antigo que se pretende substituir, o segundo parâmetro diz respeito ao caracter que irá substituir o antigo e, por fim, o último parâmetro especifica quantas ocorrências do caracter antigo se pretendem substituir, por padrão ocorre a substituição de todas as ocorrências.<br>
Para se realizar a contagem das bases de interesse usou-se o método count(), que possui 3 parâmetros. O primeiro parâmetro é obrigatório e corresponde ao caracter que se pretende realizar a contagem. O segundo (opcional) corresponde a um valor inteiro que indica a partir de que posição da string se pretende iniciar a contagem, definido por padrão definido como 0. Por fim, o último parâmetro (opcional) é um valor inteiro que vai indicar a posição do local na string onde se vai terminar a contagem, por padrão, está definido para terminar no fim da string.<br>

A função valid_strand verifica se a cadeia de nucleotidos fornecida recebe como argumento uma cadeia de DNA ou RNA  valida devolvendo um booleano que será True se a cadeia for válida. Para verificar a validade da cadeia aplica-se os métodos strip(), upper() e replace dos caracteres A, C, G, T e ' ' por uma string vazia (''). Se a cadeia fornecida for válida é esperado que após a aplicação destes métodos se obtenha uma string vazia no final.

```python
def gc_content(DNA : str) -> float:
  '''
  Função que recebe uma cadeia de DNA e calcula o teor de GC da cadeia
  
  Parameters
  -----
  DNA:str
    cadeia de DNA na qual se pretende determinar o teor de GC

  Returns
  -----
  Float
    Valor do teor de GC calculada para a cadeia de DNA fornecida

  Raises
  ----
  AssertionError
    No caso de não ser fornecida nenhuma cadeia de DNA
 
  '''
  assert len(DNA) != 0, 'A cadeia fornecida não pode ser uma string vazia' #garante que uma string vazia nao e fornecida
  DNA         = DNA.strip().upper().replace(' ','')
  C_count     = DNA.count('C')
  G_count     = DNA.count('G')
  GC_content  = (C_count/len(DNA))*100 + (G_count/len(DNA))*100
  GC_content  = round(GC_content, 2)
  return GC_content


assert gc_content  ("aaataTaTTaa") == (00.00)
assert gc_content  ('atgc AG CT' ) == (50.00)
assert gc_content  ('   atgGCT  ') == (50.00)
assert gc_content  ('GGGCGCCCCGC') == (100.00)


def valid_strand(strand : str) -> bool:
  '''
  Função que recebe uma cadeia e verifica se é uma cadeia válida de DNA
  
  Parameters
  -----
  Strand:str
    cadeia que se pretende verificar se corresponde a uma cadeia de DNA válida

  Returns
  -----
  Bool
    Retorna verdadeiro se a sequência for uma cadeia de DNA válido, caso contrário retorna falso

  '''
  return strand.strip().upper().replace('A', '').replace('C', '').replace('G', '').replace('T', '').replace(' ','') == ''

assert valid_strand          ('')
assert valid_strand          ('AAAAT')
assert valid_strand          ('AAaaaaAA')
assert valid_strand          ('AAA ttttcg t')
assert not valid_strand      ('Uu')
assert not valid_strand      ('aaaGTTT1TCCCC0')
assert not valid_strand      ('XAAABQR')
assert not valid_strand      ('AGT_CCA-GGGGCAA')
```

```python
strand = input()

if valid_strand (strand) == True:
  print('O teor de GC na cadeia fornecida é de {}%.'.format(gc_content (strand)))
else:
  print('Ocorreu um erro, verifica a cadeia fornecida')
```

A tarefa da semana 2 consiste em criar um conjunto de funções que permitam realizar a tradução de uma sequência de nucleótidos bem como a obtenção do complemento inverso de uma dada sequência de nucleotidos.

**Tradução de codões**
<div align='justify'>
Primeiramente definiou-se a função cadeia_valida, que é utilizada para verificar se a sequência de nucleotidos fornecida é uma sequência válida, pois apenas se pretende realizar a tradução de sequências de nucleótidos válidas. Esta função pode receber quer sequências de DNA, quer de RNA.<br> 
Em seguida, definui-se a função tipo_seq que é necesária para identificar se a sequência fornecida é de DNA ou de RNA de modo a escolher a tabela de tradução de codões adequada.<br> 
Depois definiu-se a função tabela_traducao, sendo que esta utiliza o resultado da função anterior e devolve a tabela de tradução de codões standard adequada para o tipo de ácido nucleico que se pretende realizar a tradução.<br>
Após isto, criou-se a função acid_nuc_2codons que é a responsável por extrair todos os codões da sequência de nucleotidos sendo que a função começa a extrair os codões apartir do início da sequência e apenas permite a obtenção dos codões para uma grelha de leitura que corresponde a grelha fornecida.<br>
Por fim, definiu-se a função traducao_acid_nuc que integra as 4 funções anteriores de modo a validar a sequência de nucleotidos, identificar o tipo da sequência de nucleotidos, escolher a tabela de tradução adequada e a obtenção de todos os codões presentes na sequência. Após isto, ocorre a tradução propiamente dita, onde os codões são substituidos pelo respetivo aminoacido e adicionados a variável seq_aa que no fim vai ser retornada e corresponde à sequência de aminoacidos obtida para a sequência de nucleotidos.<br>

**Complemento inverso**
<div align='justify'>
A funcão complemento_inverso, vai utilizar uma sequência de DNA e devolver o seu complemento inverso. Para isso utiliza a função tipo_seq que vai validar se a sequência de nucleotidos fornecida é valida e se corresponde a uma sequência de DNA. Caso isto se verifique, vai ocorrer a substituição de cada base pela sua base complementar e a inversão da sequência, isto é, a sequência passa a ser apresentada do fim para o início

```python
# tradução de codões
def cadeia_valida (strand : str) -> bool:
  '''
  Verifica se a cadeia de nucleotidos fornecida é válida

  Parameters
  ----------
  strand: str
    Uma sequência de nucleótidos
  DNA: bool
    Identifica que a sequencia de nucleotidos é de DNA

  Returns
  ---------
  bool
    Retorna verdadeiro se a cadeia fornecida for um cadeia de nucleótidos válida  

  Raises
  ---------
  AssertionError
    Se a sequencia não for fornecida no formato de string
  '''
  assert type(strand)==str, "A cadeia deve ser fornecida no formato de string"
  
  for pos,base in enumerate(strand.upper().replace(" ","")): 
    if base not in "AGCTU":
      print(f'O elemento {base}, na posição {pos+1}, não é um nucleotido válido')
      return False
  
  if "T" in strand.upper() and "U" in strand.upper():
    print('A sequencia de nucleotidos não é valida, pois possui T e U na mesma cadeia')
    return False
  
  return True

def tipo_seq(strand : str) -> str:
  '''
  Identifica o tipo de acido nucleico da sequencia fornecida

  Parameters
  ----------
  strand: str
    Uma sequência de nucleótidos

  Returns
  ---------
  str
    Retorna a classificação da cadeia de nucleótidos fornecida\n
    Por padrão, caso não esteja presente na sequencia a base uracilo, a sequência vai ser classificada como DNA
    
  Raises
  -------
  AssertionError
    Quando não é fornecida nenhuma sequência\n
    Se a sequência fornecida não for válida
  '''

  assert len(strand)!= 0, "Nenhuma cadeia foi fornecida"
  strand_valida= cadeia_valida(strand) #verificacao da validade da cadeia
  assert strand_valida,"A cadeia fornecida não é válida"

  if "U" in strand.upper(): return "RNA"
  else: return "DNA"
   
def tabela_traducao(tipo_acid_nuc : str) -> dict[str:str]:
  '''
  Devolve a respetiva tabela de tradução para o acido nucleico fornecido

  Parameters
  ----------
  tipo_acid_nuc: str
    Classificacao do acido nucleico. Apenas pode assumir o valor de DNA ou RNA
  
  Returns
  ---------
  dict
    Devolve a tabela de tradução STANDARD para o respetivo acido nucleico

  Raises
  ---------
  AssertionError
    Quando o tipo de acido nucleico fornecido nao e valido
  '''

  assert tipo_acid_nuc.upper() in ['DNA','RNA'], f'O valor {tipo_acid_nuc} não é um tipo de acido nucleico valido'

  if tipo_acid_nuc.upper()=="RNA":
    RNA_codon_table={
     "UUU":"F", "CUU":"L", "AUU":"I", "GUU":"V", "UUC":"F", "CUC":"L", "AUC":"I", "GUC":"V",
     "UUA":"L", "CUA":"L", "AUA":"I", "GUA":"V", "UUG":"L", "CUG":"L", "AUG":"M", "GUG":"V",
     "UCU":"S", "CCU":"P", "ACU":"T", "GCU":"A", "UCC":"S", "CCC":"P", "ACC":"T", "GCC":"A",
     "UCA":"S", "CCA":"P", "ACA":"T", "GCA":"A", "UCG":"S", "CCG":"P", "ACG":"T", "GCG":"A",
     "UAU":"Y", "CAU":"H", "AAU":"N", "GAU":"D", "UAC":"Y", "CAC":"H", "AAC":"N", "GAC":"D",
     "UAA":"_", "CAA":"Q", "AAA":"K", "GAA":"E", "UAG":"_", "CAG":"Q", "AAG":"K", "GAG":"E",
     "UGU":"C", "CGU":"R", "AGU":"S", "GGU":"G", "UGC":"C", "CGC":"R", "AGC":"S", "GGC":"G",
     "UGA":"_", "CGA":"R", "AGA":"R", "GGA":"G", "UGG":"W", "CGG":"R", "AGG":"R", "GGG":"G"}
    return RNA_codon_table
  
  else:
    DNA_codon_table={
      "ATA":"I", "ATC":"I", "ATT":"I", "ATG":"M", "ACA":"T", "ACC":"T", "ACG":"T", "ACT":"T",
      "AAC":"N", "AAT":"N", "AAA":"K", "AAG":"K", "AGC":"S", "AGT":"S", "AGA":"R", "AGG":"R",                
      "CTA":"L", "CTC":"L", "CTG":"L", "CTT":"L", "CCA":"P", "CCC":"P", "CCG":"P", "CCT":"P",
      "CAC":"H", "CAT":"H", "CAA":"Q", "CAG":"Q", "CGA":"R", "CGC":"R", "CGG":"R", "CGT":"R",
      "GTA":"V", "GTC":'V', "GTG":'V', "GTT":"V", "GCA":"A", "GCC":"A", "GCG":"A", "GCT":"A",
      "GAC":"D", "GAT":"D", "GAA":"E", "GAG":"E", "GGA":"G", "GGC":"G", "GGG":"G", "GGT":"G",
      "TCA":"S", "TCC":"S", "TCG":"S", "TCT":"S", "TTC":"F", "TTT":"F", "TTA":"L", "TTG":"L",
      "TAC":"Y", "TAT":"Y", "TAA":"_", "TAG":"_", "TGC":"C", "TGT":"C", "TGA":"_", "TGG":"W"}
    return DNA_codon_table

def acid_nuc_2codons(strand : str) -> list[str]:
  '''
  Devolve todos os codões para uma determinada sequencia de nucleotidos
  
  Parameters
  ----------
  Strand: str
    Uma sequência válida de um acido nucleico

  Returns
  ---------
  codoes: list
    Devolve uma lista de strings que contem todos os codões

  Raises
  -------
  AssertionError
    Quando a sequencia fornecida não é valida\n
    Caso a sequência possua tamanho inferior a 3
  '''
  cad_valida=cadeia_valida(strand)
  assert cad_valida, 'A cadeia fornecida não é valida'
  assert len(strand)>=3, 'A cadeia fornecida apenas possui {} bases. São necessárias pelo menos 3 bases para formar codões'.format(len(strand))
  codoes=[]
  for I in range(0,len(strand)-2,3):
    codao=strand[I:I+3]
    codoes.append(codao.upper())
  return codoes

def traducao_acid_nuc(strand : str) -> str:
  '''
  Realiza a traducao da sequencia de nucleotidos para uma sequencia de aminoacidos

  Parameters
  ----------
  Strand: str
    Sequência de um acido nucleico

  Returns
  ---------
  seq_aa: str
    Sequência de aminoácidos resultantes da tradução

  Raises
  -------
  AssertionError
    Se a cadeia fornecida não for uma sequencia de nucleotidos valida\n
    Quando a cadeia fornecida é uma string vazia\n
    Caso a cadeia fornecida possua tamanho inferior a 3
  '''

  tipo_acid_nuc=tipo_seq(strand) #classificacao do acido nucleico
  codoes= acid_nuc_2codons(strand) #gerarcao de todos os codoes da sequencia
  tabela_codoes=tabela_traducao(tipo_acid_nuc) #obtencao da tabela de traducao 
  seq_aa=""

  for codao in codoes: seq_aa+=tabela_codoes[codao] #traducao dos codoes
  return seq_aa
```

```python
# complemento inverso
def complemento_inverso (strand : str) -> str:
  '''
  Determina a cadeia complementar inversa de uma sequencia de DNA 

  Parameters
  ----------
  Strand: str
    Sequência de nucleotidos

  Returns
  ---------
  complementar_inv: str
    Complementar inverso da sequência de DNA fornecida

  Raises
  -------
  AssertionError
    Se não for fornecida nenhuma sequencia\n
    Caso a sequencia de nucleotidos não seja valida\n
    Quando a sequencia é de RNA
  '''
  tip_seq=tipo_seq(strand) #classificacao do acido nucleico e verificacao da sua validade
  assert tip_seq=="DNA", "A sequência fornecida tem de ser de DNA"
  complementar=""
  
  DNA_bases_complementares={"A":"T","T":"A","G":"C","C":"G"}
  for base in strand.upper(): complementar+=DNA_bases_complementares[base]
  
  return complementar[::-1]
```

## Tarefa 3 - Tradução do DNA, ORF (open reading frames) e extração de todas as proteínas possíveis

<div align='justify'>

Nesta tarefa é proposto que a partir de uma sequência de DNA se proceda à extração de todas as proteínas que podem ser codificadas por uma sequência de DNA nas várias reading frames existentes. Para se realizar a extração de todas as proteínas que se encontram codificadas numa sequência de DNA é necessário determinar os codões de cada reading frame, realizar a tradução dos codões de acordo com uma tabela de tradução de codões considerando que podem existir vários codões de iniciação para cada tabela de tradução utilizada, extrair, para cada sequência de aminoacidos resultante, todas as possíveis proteínas e, por fim, apresentar as diversas proteínas que é possível obter da sequência de DNA.<br>

Para atingir este objetivo é necessário definir as seguintes funções:<br>
Uma função que valide a sequência de DNA fornecida e uma função que permita obter o complementar inverso do DNA em estudo. Estas funções foram definidas na tarefa anterior (tipo_sep e complemento_inverso)<br>
Para se obter os codões de uma determinada reading frame criou-se a função rf_codoes, que foi posteriormente integrada na função all_rfs que permite obter os codões para todas as reading frames possíveis.<br>
Definiu-se a função codon_table de modo a poder selecionar a tablela de tradução adequada, isto é, permitir ajustar a tabela de tradução conforme o organismo da qual proveem o DNA. A função start_codon foi definida para verificar se o codão que se pretende usar para iniciar a tradução de proteínas constitui uma alternativa válida para a tabela de tradução utilizada enquanto a função stop_codons foi definida com o mesmo inuito, mas para os codões de stop.
Em seguida criou-se a função traducao_rf de modo a poder realizar a tradução dos codões para cada reading frame. Por fim, definiu-se a função find_prots que encontra as possíveis proteínas existentes na sequência de aminoacidos que resulta da tradução.<br>
A função all_prots integra todas as funções abordadas anteriormente e devolve um conjunto de proteínas que corresponde a todas as proteínas distinas que são codificadas.
</div>

```python
def rf_codoes(strand : str, reading_frame = 0) -> list[str]:
  '''
  Extrai todos os codões de uma sequência de DNA para uma determinada reading frame

  Parameters
  -----
  strand: str
    Cadeia de DNA da qual se pretende realizar a extração dos codões para um determinada ORF
    
  reading_frame: int
    Corresponde à reanding frame que se pretende utilizar. Por padrão assume o valor 0, mas pode também assumir o valor 1 e 2

  Returns
  -------
  codoes: list[str]
    Corresponde aos vários codões para a reading frame indicada

  Raises
  ------
  AssertionError
    Se a variavél strand não for do tipo string\n
    Caso a cadeia fornecida não seja válida e/ou não seja uma cadeia de DNA\n
    Quando o valor do parametro reading_frame não está entre 0 e 2 inclusive e/ou não é um inteiro
  '''
  assert type(strand)==str #forcar a variavel strand a ser uma string
  assert tipo_seq(strand)=='DNA','A sequência fornecida deve ser de DNA' #funcao da semana anterior
  assert reading_frame >=0 and reading_frame<=2 and type(reading_frame)==int,"A reading frame fornecida deve ser um inteiro entre 0 e 2"
    
  codoes=[strand[indice:indice+3].upper() for indice in range(reading_frame,len(strand)-2,3)]
    
  return codoes

def all_rfs(strand : str) -> list[list[str]] :
  '''
  Devolve todos os codões para cada reading frame possível da sequencia de DNA indicada

  Parameters
  ----------
  strand: str
    Cadeia de DNA em estudo

  Returns
  ----------
  rfs_codoes: list[list[str]]
    Corresponde a todos os codões possíveis para as várias reading frames

  Raises
  -------
  AssertionError
    Se a sequencia fornecida não for válida e/ou não for uma sequência de DNA\n
    Quando a strand é uma string vazia
  '''
  assert type(strand)==str
  inv=complemento_inverso(strand) #funcao da semana passada
  rfs_codoes=[rf_codoes(strand,I)for I in range(3)] + [rf_codoes(inv,I) for I in range(3)]
  return rfs_codoes

def codon_table(genetic_code : str, nomes = False) -> dict[str,str]:
  '''
  Devolve a tabela de tradução adequada para o codigo genético indicado

  Parameters
  ----------
  genetic_code: str
    String que identifica o codigo genético a utilizar. Quando assume o valor de 'Vertebrate Mitochondrial' a
    tabela fornecida é a da mitocondria do humano

  nomes: bool
    Por padrao encontra-se definido como False. Pode também assumir o valor de True
    
  Returns
  ---------
  dict
    Quando o parâmetros nomes assume o booleano False, devolve a tabela de tradução de codões para o 
    respetivo codigo genético na forma de dicionario

  codigos: list[str]
    Quando o parâmetro nomes assume o valor True é devolvida uma lista com todos os códigos genéticos
    que possuem uma tabela de tradução associada

  Raises
  -------
  AssertionError
    Caso o codigo genético não seja fornecido em formato de string\n
    Quando o código genético fornecido não possui nenhuma tabela associada
  '''
  codigos=["Standard","Vertebrate Mitochondrial","Bacterial"]
  
  if nomes:
    return codigos

  else:
    assert type(genetic_code)==str,'O parâmetro genetic_code apenas pode ser do tipo string'
    assert genetic_code in codigos,f'O codigo genetico {genetic_code} não está disponível'
    
    if genetic_code=="Standard":
      standard_table={
        "ATA":"I", "ATC":"I", "ATT":"I", "ATG":"M", "ACA":"T", "ACC":"T", "ACG":"T", "ACT":"T",
        "AAC":"N", "AAT":"N", "AAA":"K", "AAG":"K", "AGC":"S", "AGT":"S", "AGA":"R", "AGG":"R",                
        "CTA":"L", "CTC":"L", "CTG":"L", "CTT":"L", "CCA":"P", "CCC":"P", "CCG":"P", "CCT":"P",
        "CAC":"H", "CAT":"H", "CAA":"Q", "CAG":"Q", "CGA":"R", "CGC":"R", "CGG":"R", "CGT":"R",
        "GTA":"V", "GTC":'V', "GTG":'V', "GTT":"V", "GCA":"A", "GCC":"A", "GCG":"A", "GCT":"A",
        "GAC":"D", "GAT":"D", "GAA":"E", "GAG":"E", "GGA":"G", "GGC":"G", "GGG":"G", "GGT":"G",
        "TCA":"S", "TCC":"S", "TCG":"S", "TCT":"S", "TTC":"F", "TTT":"F", "TTA":"L", "TTG":"L",
        "TAC":"Y", "TAT":"Y", "TAA":"_", "TAG":"_", "TGC":"C", "TGT":"C", "TGA":"_", "TGG":"W"}
      return standard_table
  
    elif genetic_code == "Vertebrate Mitochondrial":
      homo_verte_mito_table={
        "TTT": "F", "TCT": "S", "TAT": "Y", "TGT": "C", "TTC": "F", "TCC": "S", "TAC": "Y", "TGC": "C", 
        "TTA": "L", "TCA": "S", "TAA": "_", "TGA": "W", "TTG": "L", "TCG": "S", "TAG": "_", "TGG": "W",
        "CTT": "L", "CCT": "P", "CAT": "H", "CGT": "R", "CTC": "L", "CCC": "P", "CAC": "H", "CGC": "R",
        "CTA": "L", "CCA": "P", "CAA": "Q", "CGA": "R", "CTG": "L", "CCG": "P", "CAG": "Q", "CGG": "R",
        "ATT": "I", "ACT": "T", "AAT": "N", "AGT": "S", "ATC": "I", "ACC": "T", "AAC": "N", "AGC": "S",
        "ATA": "M", "ACA": "T", "AAA": "K", "AGA": "_", "ATG": "M", "ACG": "T", "AAG": "K", "AGG": "_",
        "GTT": "V", "GCT": "A", "GAT": "D", "GGT": "G", "GTC": "V", "GCC": "A", "GAC": "D", "GGC": "G",
        "GTA": "V", "GCA": "A", "GAA": "E", "GGA": "G", "GTG": "V", "GCG": "A", "GAG": "E", "GGG": "G"}
      return homo_verte_mito_table
  
    elif genetic_code=="Bacterial":
      bacterial_table={
        "TTT": "F", "TCT": "S", "TAT": "Y", "TGT": "C", "TTC": "F", "TCC": "S", "TAC": "Y", "TGC": "C",
        "TTA": "L", "TCA": "S", "TAA": "_", "TGA": "_", "TTG": "L", "TCG": "S", "TAG": "_", "TGG": "W",
        "CTT": "L", "CCT": "P", "CAT": "H", "CGT": "R", "CTC": "L", "CCC": "P", "CAC": "H", "CGC": "R",
        "CTA": "L", "CCA": "P", "CAA": "Q", "CGA": "R", "CTG": "L", "CCG": "P", "CAG": "Q", "CGG": "R",
        "ATT": "I", "ACT": "T", "AAT": "N", "AGT": "S", "ATC": "I", "ACC": "T", "AAC": "N", "AGC": "S",
        "ATA": "I", "ACA": "T", "AAA": "K", "AGA": "R", "ATG": "M", "ACG": "T", "AAG": "K", "AGG": "R",
        "GTT": "V", "GCT": "A", "GAT": "D", "GGT": "G", "GTC": "V", "GCC": "A", "GAC": "D", "GGC": "G", 
        "GTA": "V", "GCA": "A", "GAA": "E", "GGA": "G", "GTG": "V", "GCG": "A", "GAG": "E", "GGG": "G"}
      return bacterial_table
    
def start_codons(genetic_code : str) -> list:
  '''
  Devolve todos os codões de iniciação possíveis para o codigo genético indicado

  Parameters
  ----------
  genetic_code: str
    String que identifica o codigo genético a utilizar. Quando assume o valor de 'Vertebrate Mitochondrial' devolve os codões de iniciação
    para a mitocondria do humano
    
  Returns
  ----------
  start: list[str]
    Lista que contem todos os codões de iniciação possíveis para determinado código genético

  Raises
  -------
  AssertionError
    Caso o codigo genético não seja fornecido em formato de string\n
    Quando o código genético fornecido não possui nenhuma tabela associada
  '''
  codigos=["Standard","Vertebrate Mitochondrial","Bacterial"]
  assert type(genetic_code)==str,'O parâmetro genetic_code apenas pode ser do tipo string'
  assert genetic_code in codigos,f'O codigo genetico {genetic_code} não está disponível'
    
  if genetic_code=="Standard":
    start=['ATG','CTG','TTG']
    return start
  
  elif genetic_code == "Vertebrate Mitochondrial":
    start=['ATG','ATT']
    return start
    
  elif genetic_code=="Bacterial":
    start=['ATG','ATA','ATC','ATT','CTG','TTG','GTC']
    return start
  
def stop_codons(genetic_code : str) -> list[str]:
  '''
  Devolve todos os codões de stop possíveis para o codigo genético indicado

  Parameters
  ----------
  genetic_code: str
    String que identifica o codigo genético a utilizar.
    
  Returns
  ----------
  stop: list[str]
    Lista que contem todos os codões de stop possíveis para um determinado código genético

  Raises
  -------
  AssertionError
    Caso o codigo genético não seja fornecido em formato de string\n
    Quando o código genético fornecido não possui nenhuma tabela associada
  '''
  codigos=["Standard","Vertebrate Mitochondrial","Bacterial"]
  assert type(genetic_code)==str,'O parâmetro genetic_code apenas pode ser do tipo string'
  assert genetic_code in codigos,f'O codigo genetico {genetic_code} não está disponível'
    
  if genetic_code=="Standard":
    stop=['TAA','TAG','TGA']
    return stop
  
  elif genetic_code == "Vertebrate Mitochondrial":
    stop=['TAA','TAG','AGA','AGG']
    return stop
    
  elif genetic_code=="Bacterial":
    stop=['TAA','TAG','TGA']
    return stop

def traducao_rf(all_rf_codoes : list[list[str]], genetic_code = "Standard", start_codon = 'ATG') -> list[str]:
  '''
  Devolve uma sequência de aminoácidos para cada reading frame

  Parameters
  ----------
  rf_codoes: list[list[str]]
    Lista que contêm todos os codões em formato de string para uma determinada reading frame

  genetic_code: str
    Indica que tabela de tradução deve ser utilizada. Por padrão utiliza a tabela standard

  start_codon: str
    Indica qual o codão de iniciação a ser utilizada. Utiliza o codão ATG como default
 
  Returns
  ---------
  seq_aas: list[str]
    Sequência de aminoácidos resultantes da tradução para cada reading frame

  Raises
  -------
  AssertionError
    Caso o codão de start fornecido não seja um codão válido ou não seja fornecido no formato de string\n
    Se o código genético não for fornecido em formato de string ou não possuir tabela de tradução associada
  '''
  trans_table=codon_table(genetic_code)
  assert type(start_codon)==str,'O codão de iniciação deve ser do tipo string'
  start=start_codons(genetic_code)
  assert start_codon.upper() in start,f'O codão de iniciação {start_codon} não ocorre no codigo genético {genetic_code}'
  stop=stop_codons(genetic_code)
  estado=False
  seq_aas=[]
  
  for rf in all_rf_codoes:
    seq_aa=''
    for codao in rf:
      if estado:
       seq_aa+=trans_table[codao.upper()]
       if codao.upper() in stop:
         estado=False
         continue

      if not estado:
        if codao.upper()==start_codon.upper():
         seq_aa+='M'
         estado=True
        
        else:seq_aa+=trans_table[codao.upper()]

    seq_aas.append(seq_aa)

  return seq_aas

def find_prots(seq_aa : str) -> list[str]:
  '''
  Devolve todos as proteínas hipotéticas presentes em uma sequência de aminoácidos

  Parameters
  ----------
  seq_aa: str
    Sequência de aminoacidos da qual se prende extrair todas as proteinas existentes na sequencia
 
  Returns
  ---------
  seq_aas: list[str]
      Sequência de aminoácidos resultantes da tradução para cada reading frame

  Raises
  -------
  AssertionError
    Quando a sequencia fornecida não está no formato de string\n
    Caso o aminoacido presente na sequencia não seja um aminoácido válido
  '''
  assert type(seq_aa)==str,'A sequencia de aminoacidos deve ser do tipo str'
  alfabeto='ACDEFGHIKLMNPQRSTVWY_'
  prots=[]
  prot=''
  estado=False

  for aa in seq_aa.upper():
    assert aa in alfabeto,f'O aminoacido {aa} não esta presente no alfabeto dos aminoacidos'
    if estado:
      if aa=="_":
        estado=False
        prots.append(prot)
        prot=''
        continue
      else:prot+=aa

    else:
      if aa=='M':
        prot+=aa
        estado=True

  return prots

def all_prots(strand : str, gen_code = 'Standard', start = 'ATG') -> set:
  '''
  Devolve todos as proteínas possíveis que são codificas pela cadeia fornecida em todas as reading frames

  Parameters
  ----------
  strand: str
    Cadeia de DNA da qual se pretende extrair todas as proteínas existentes nas 6 reading frames

  gen_code: str
    Corresponde ao código genético a utilizar para realizar a tradução

  start: str
    Codão de iniciação a utilizar para realizar a tradução dos codões

  Returns
  ---------
  todas_proteinas: set()
    Corresonde ao conjunto de todas as proteínas distintas que podem ser codificadas pela sequênica nas 6
    reading frames possíveis

  Raises
  -------
  AssertionError
    Se a sequência fornecida não for válida e/ou não for uma sequência de DNA\n
    Quando a strand corresponde a uma string vazia\n
    Caso o codão start fornecido não seja válido ou não seja fornecido em formato de string\n
    Se o código não for fornecido em formato de string ou não possua tablea de tradução associada\n
  '''
  rfs_cod=all_rfs(strand) #obtenção dos codoes para todas as reading frames
  rfs_seq_aa=traducao_rf(rfs_cod,genetic_code=gen_code,start_codon=start) #traducao dos codoes de cada reading frame
  todas_proteinas=set()
  for seq in rfs_seq_aa: #navegacao entre as várias sequencias de aminoacidos
    prots=find_prots(seq) #obtenção de todas as proteinas existens na sequencia de aminoacidos
    for prot in prots:
      todas_proteinas|=set([prot])

  if len(todas_proteinas)==0:
    print('A sequência não codifica nenhuma proteína')
    return todas_proteinas
    
  else:
    print(f'A sequência codifica {len(todas_proteinas)} proteinas diferentes em todas as reading frames possíveis')
    return todas_proteinas
```

## Tarefa 4 - Alinhamento através de progração dinâmica/ recursividade

<div align='justify'>

Na tarefa da semana 4, é proposto a resolução <br>
 score_basic - Esta função utiliza recursão para calcular o escore de alinhamento básico entre duas sequências.
 memoize - A função memoize armazena os resultados previamente calculados em um dicionário de memória.
    Sempre que a função memoizada é chamada com determinados argumentos, verifica-se se esses argumentos já foram
    calculados anteriormente. Se sim, o resultado armazenado é retornado; caso contrário, a função original é chamada
    para calcular o resultado, que é então armazenado na memória para uso futuro.
 score - Esta função utiliza recursão para calcular o escore de alinhamento entre duas sequências.

```python
#Sem memoizing
def score_basic(s1 :str, s2 : str, g : int= -4, match : int = 2, mismatch : int = -1) -> int:
    """
    Calcula o score de alinhamento básico entre duas sequências utilizando a recursão.
        
    Parameters
    ----------
    s1 : str
      Primeira sequência.

    s2 : str
      Segunda sequência.

    g : int
      Valor a ser utilizado para penalizar os espaçamentos. Por padrão, considera-se o valor de penalização como -4.

    match : int
      Valor a ser utilizado quando os caracteres são iguais. Por padrão, é 2.

    mismatch : int
      Valor a ser utilizado quando os caracteres são diferentes. Por padrão, é -1.

    Returns
    -------
    int
      O score de alinhamento básico entre as duas sequências.

    Raises
    ------
    AssetionError
      Gap penalty tem de ser valor inteiro e negativo.
      Mismatch tem de ser valor inteiro e negativo.
      Match tem de ser valor inteiro e positiv
    """
    assert g <0 and isinstance(g,int) , "Gap penalty tem de ser valor inteiro e negativo "
    assert mismatch <0 and isinstance(mismatch,int) , "Mismatch tem de ser valor inteiro e negativo"
    assert match > 0 and isinstance(match,int) , "Match tem de ser valor inteiro e positivo"
    
    subst = lambda X, Y: match if X == Y else mismatch
    if s1 and s2:
        return max(
            score_basic(s1[:-1], s2[:-1]) + subst(s1[-1], s2[-1]),
            score_basic(s1[:-1], s2) + g,
            score_basic(s1, s2[:-1]) + g
            )
    else: return max(len(s1), len(s2)) * g

score_basic("HGWAG","PHSWG")
```

```python
#Com memoizing usando um decorador
def memoize(func):
    """
    Memoiza uma função, armazenando os resultados previamente calculados para evitar recálculos.
        
    Parameters
    ----------
    func : function
      A função a ser memoizada.

    Returns
    -------
    function
      A função memoizada.
    """
    memoria = {}
    def organizador(*positional_args, **keyword_args):
        if positional_args not in memoria:
            memoria[positional_args] = func(*positional_args, **keyword_args)
        return memoria[positional_args]
    return organizador

@memoize
def score_memo(s1 : str, s2 : str, g : int = -4, match : int=2, mismatch : int = -1, indent : int =0, check_indent : bool = False) -> int:
    """
    Calcula o score de alinhamento entre duas sequências utilizando a recursão.
        
    Parameters
    ----------
    s1 : str
      Primeira sequência.

    s2 : str
      Segunda sequência.

    g : int
      Valor a ser utilizado para penalizar os espaçamentos. Por padrão, considera-se o valor de penalização como -4.

    match : int
      Valor a ser utilizado quando os caracteres são iguais. Por padrão, é 2.

    mismatch : int
      Valor a ser utilizado quando os caracteres são diferentes. Por padrão, é -1.

    indent : int
      Nível de indentação para exibição de informações (usado apenas se check_indent=True). Por padrão, é 0.

    check_indent : bool
      Indica se deve imprimir informações de indentação durante o processo de recursão. Por padrão, é False.

    Returns
    -------
    int
      O score de alinhamento entre as sequências.

    Raises
    ------
    AssetionError
      Gap penalty tem de ser valor inteiro e negativo.
      Mismatch tem de ser valor inteiro e negativo.
      Match tem de ser valor inteiro e positivo.
    """
    assert g <0 and isinstance(g,int) , "Gap penalty tem de ser valor inteiro e negativo "
    assert mismatch <0 and isinstance(mismatch,int) , "Mismatch tem de ser valor inteiro e negativo"
    assert match > 0 and isinstance(match,int) , "Match tem de ser valor inteiro e positivo"
    subst = lambda X, Y: match if X == Y else mismatch

    
    if s1 and s2:
        val = max(
            score_memo(s1[:-1], s2[:-1], g, match, mismatch, indent=indent + 1, check_indent=check_indent) + subst(s1[-1], s2[-1]),
            score_memo(s1[:-1], s2,      g, match, mismatch, indent=indent + 1, check_indent=check_indent) + g,
            score_memo(s1,      s2[:-1], g, match, mismatch, indent=indent + 1, check_indent=check_indent) + g
        )
        if check_indent==True:
            print("     " * indent + f"Alinhamento recursivo de s1: '{s1}' com s2: '{s2}' com indent: {indent}")
        return val
    else:
        return max(len(s1), len(s2)) * g
    
score_memo("HGWAG","PHSWG", check_indent=True)
```

```python
def validar_prot(seq : str)->bool:
  """
  Esta função recebe uma sequência de aminoácidos e realiza as seguintes operações:
      1. Verifica se a sequência é do tipo string.
      2. Verifica se a sequência não é vazia.
      3. Verifica se a sequência contém apenas os aminoácidos permitidos (A, C, D, E, F, G, H, I, K, L, M, N, P, Q, R, S, T, V, W, Y, _).

  Parameters
  ----------
  seq : str
    A sequência de aminoácidos que será validada.

  Returns
  -------
  bool
    Retorna True se a sequência de aminoacidos for válida, caso contrário, retorna False.

  Raises
  ------
  AssertionError
    - Se a sequência não for fornecida no formato de string.
    - Se a sequência fornecida for vazia.
  """
  assert type(seq)==str,'A sequência não foi fornecida em formato de string'
  assert len(seq) !=0 ,'Não foi fornecida nenhuma sequência'
  lista_aminoacidos = ['A','C','D','E','F','G','H','I','K','L','M','N','P','Q','R','S','T','V','W','Y','_']
  for aa in seq.upper():
      if aa not in lista_aminoacidos:
          return False
    
  return True

def formatar(x : int or str) -> str:
  """
  Esta função recebe um valor x e realiza as seguintes operações:
    1. Se x for do tipo inteiro, formata o valor como uma string com pelo menos 3 dígitos, preenchendo com espaços à esquerda se necessário.
    2. Se x não for do tipo inteiro, formata o valor como uma string com pelo menos 3 caracteres, preenchendo com espaços à esquerda se necessário.

  Parameters
  ----------
  x : int or str
    O valor que será formatado.

  Returns
  -------
  str
    O valor formatado.

  Raises
  ------
  ValueError
    Se x não for do tipo inteiro nem string.
  """
  if type(x) is int:
      return f"{x:>3d}"
  else:
      return f"{x:>3}"
    
def print_matrix(S1 : str, S2 : str, M : list) -> None:
  """
  Esta função recebe duas sequências (S1 e S2) e uma matriz (M) e realiza as seguintes operações:
    1. Imprime uma representação formatada da matriz.
    2. Utiliza a função 'formatar' para formatar os valores da matriz.

  Parameters
  ----------
  S1 : str
    A primeira sequência.

  S2 : str
    A segunda sequência.

  M : list
    A matriz a ser impressa.

  Returns
  -------
  None
    A função imprime a matriz formatada.

  Raises
  ------
  ValueError
    Se S1 ou S2 não forem fornecidos como strings.
    Se M não for uma lista válida.
  """
  print(*list(" " + S1), sep="   ")
  for x2, L in zip(S2,M): #x2 caracter que vem da string S2
      LL=[formatar(x) for x in L]
      print(x2,*LL)
  print()

def score_sub_acid_nuc(x1 : str, x2 : str, match : int, mismatch : int) -> int:
  """
  Esta função recebe dois caracteres x1 e x2, representando nucleotídos ou aminoácidos, e realiza as seguintes operações:
    - Retorna o valor de correspondência (match) se x1 e x2 são iguais.
    - Retorna o valor de não correspondência (mismatch) se x1 e x2 são diferentes.

  Parameters
  ----------
  x1 : str
    O primeiro caractere (nucleotídeo ou aminoácido).

  x2 : str
    O segundo caractere (nucleotídeo ou aminoácido).

  match : int
    O valor atribuído quando x1 e x2 são iguais. O valor padrão para match é 2.

  mismatch : int
    O valor atribuído quando x1 e x2 são diferentes. O valor padrão para mismatch é -1.

  Returns
  -------
  int
    O valor de correspondência (match) se x1 e x2 são iguais, caso contrário, o valor de não correspondência (mismatch).
  """
  if x1==x2:return match
  else: return mismatch

def score_subst_prot(nome_matriz : str, aa1 : str, aa2 : str) -> int:
  """
  Esta função abre e lê um destes ficheiros de texto: "BLOSSOM62.txt", "BLOSSOM80.txt", "PAM30.txt", "PAM70.txt" 
  retirando a matriz de substituição do ficheiro como uma matriz bidimensional.
  Depois navega na matriz de substituição obtida e retorna o score correspondente ao alinhamento de dois dos seguintes caracteres: 
  'A','C','D','E','F','G','H','I','K','L','M','N','P','Q','R','S','T','V','W','Y','_',
  sendo que as letras representam os aminoácidos e o caracter '_' representa o codão stop.
    
  Parameters
  ----------
  nome_matriz : str
    Nome da matriz de substituição que se pretende que seja lida e utilizada para fornecer os scores de substituição

  aa1 : str
    Aminoácido que vai ser procurado ao longo da primeira coluna da matriz de substituição

  aa2 : str
    Aminoácido que vai ser procurado ao longo da primeira linha da matriz de substituição

  Returns
  -------
  score : int
    Score obtido para o alinhamento dos dois aminoácidos na matriz de substituição

  Raises
  ------
  AssertionError
    - Quando o nome da matriz fornecido não corresponde a uma das matrizes aceites pela função:
    "BLOSSOM62", "BLOSSOM80", "PAM30", "PAM70"
    - Quando o caracter inserido correspondente ao 'aa1' não é válido como aminoácido ou codão stop
    - Quando o caracter inserido correspondente ao 'aa2' não é válido como aminoácido ou codão stop
  """
  assert nome_matriz.upper() in ["BLOSSOM62", "BLOSSOM80", "PAM30", "PAM70"],"A matriz de substituição escolhida é inválida, apenas aceita as matrizes 'BLOSSOM62', 'BLOSSOM80', 'PAM30', 'PAM70'"
  with open(nome_matriz.upper()+".txt", "r") as file:
      matriz = []
      for line in file:
          line = line.strip()
          linha = []
          valor = ""
          for elem in line:
              if elem == " ":
                  if len(valor) != 0:
                      linha.append(valor)
                      valor = ""
              else:
                  valor += elem
          linha.append(valor)
          matriz.append(linha)
  aa1, aa2 = aa1.upper(), aa2.upper()
  lista_aminoacidos = ['A','R','N','D','C','Q','E','G','H','I','L','K','M','F','P','S','T','W','Y','V','B','J','Z','X','_']
  assert aa1 in lista_aminoacidos,f"O caracter {aa1} não é válido como aminoácido ou codão stop"
  assert aa2 in lista_aminoacidos,f"O caracter {aa2} não é válido como aminoácido ou codão stop"
  coluna, linha = 0, 0
  for i in range(1,len(matriz)):
      if aa1 == matriz[i][0]:
          linha = i
          break
  for pos, elem in enumerate(matriz[0]):
      if elem == aa2:
          coluna = pos
          break
  score = matriz[linha][coluna]
  return int(score)

def reconstrucao(seq1 : str, seq2 : str, trace : list[list],matrizes=True) -> tuple[str,str]:
  """
  Esta função recebe duas sequências seq1 e seq2, juntamente com uma matriz de rastreamento trace,
  e realiza as seguintes operações:
    - Reconstrói as sequências alinhadas a partir da matriz de rastreamento.

  Parameters
  ----------
  seq1 : str
    A primeira sequência.

  seq2 : str
    A segunda sequência.

  trace : list
    A matriz de rastreamento utilizada para alinhar as sequências.
  
  matrizes: bool
    Indica se se pretende realizar o print das matrizes trace, score e das sequências alinhadas
    
  Returns
  -------
  tuple of str
    Um tuplo contendo as sequências alinhadas (aligned_seq1, aligned_seq2).
  """
  aligned_seq1 = ''
  aligned_seq2 = ''
  linha=len(seq1)-1
  coluna=len(seq2)-1

  while trace[linha][coluna]!=0:
      if trace[linha][coluna] == '↖':
          aligned_seq1 = seq1[linha] + aligned_seq1
          aligned_seq2 = seq2[coluna] + aligned_seq2
          linha -= 1
          coluna -= 1
      elif trace[linha][coluna] == '↑':
          aligned_seq1 = seq1[linha] + aligned_seq1
          aligned_seq2 = '-' + aligned_seq2
          linha -= 1
      elif trace[linha][coluna] == '←':
          aligned_seq1 = '-' + aligned_seq1
          aligned_seq2 = seq2[coluna] + aligned_seq2
          coluna-= 1
  
  if matrizes:
    print(aligned_seq1,aligned_seq2,sep='\n')
  return (aligned_seq1,aligned_seq2)

def needleman_wunsch(seq1 : str, seq2 : str, seq_tipo : str, gap : int=-4, matriz : str='Blossom62', match : int=2, mismatch : int=-1, matrizes=True ):
  """
  Realiza o alinhamento global de duas sequências recorrendo o algoritmo Needleman-Wunsch.

  Parameters
  ----------
  seq1 : str
    Primeira sequência

  seq2 : str
    Segunda sequência

  seq_tipo : str
    Identifica o tipo das sequências que se pretende alinhar

  gap : int
    Valor a utilizar para penalizar os espaçamentos. Por padrão considera-se o valor de penalização como -4

  matriz : str
    Matriz de substituição de aminoacidos a utilizar para realizar o alinhamento de proteínas

  match : int
    Valor a utilizar para quando as duas bases nucleicas são iguais. Por default é 2. Apenas para o alinhamento de DNA/RNA

  mismatch : int
    Valor a utilizar para quando as duas bases nucleicas são diferentes. Por default é -1. Apenas para o alinhamento de DNA/RNA

  matrizes : bool
    Indica se se prende realizar o print das matrizes de score e de traceback bem como as sequências alinhadas

  Returns
  ----------
  Tuple[str,[Tuple[str,str]]
    Devolve um tuplo com o score com um tuplo do alinhamento ótimo de cada sequência

  Raises
  ------
  AssertionError
     - Se seq1 e seq2 não forem do tipo string
     - Se seq 1 ou seq2 forem strings vazias
     - Se seq_tipo não for 'DNA', 'RNA' ou 'PROTEIN'.
     - Se uma das sequencias fornecidas não é um acido nucleico válido quando seq_tipo assume o valor DNA ou RNA.
     - Se seq_tipo for 'PROTEIN' e as sequências não forem proteínas válidas.
  """
  assert type(seq1)==str and type(seq2)==str
  assert seq_tipo.upper() in ['DNA','RNA','PROTEIN'], 'O tipo de sequência fornecido não é valido, apenas pode assumir o valor de DNA, RNA ou protein'
    
  if seq_tipo.upper()=='DNA' or seq_tipo.upper()=='RNA':
      assert tipo_seq(seq1) in ['DNA','RNA'], 'A sequencia 1 não é um acido nucleico'
      assert tipo_seq(seq2) in ['DNA','RNA'], 'A sequencia 2 não é um acido nucleico'
      assert tipo_seq(seq1)==tipo_seq(seq2), 'As cadeias fornecidas não são do mesmo tipo'

  else:
      assert validar_prot(seq1), 'A sequencia 1 não é uma proteina'
      assert validar_prot(seq2), 'A sequencia 2 não é uma proteina'

  seq1='-'+seq1
  seq2='-'+seq2
  nrows = len(seq1)
  ncols = len(seq2)

  score_matrix=[[0]*ncols for linhas in range(nrows)] # iniciar a matriz score apenas com 0
  score_matrix[0]=[c*gap for c in range(ncols)] #prenchimento da primeira linha da matriz score
  for linha in range(1, nrows):score_matrix[linha][0] = linha * gap # preenchimento da primeira coluna da matriz score

  traceback_matrix = [[0] * ncols for linhas in range(nrows)] # iniciar a matriz traceback com 0
  traceback_matrix[0] =[0 if col==0 else '←' for col in range(ncols)] # prenchimento da primeira linha da matriz traceback
  for linha in range(1,nrows): traceback_matrix[linha][0]='↑' # preenchimento da primeira coluna da matriz traceback

  if seq_tipo.upper()=='DNA' or seq_tipo.upper()=='RNA': #alinhamento de acidos nucleicos
    for linha in range(1, nrows):
      for coluna in range(1, ncols):
        diagonal = score_matrix[linha - 1][coluna - 1] + score_sub_acid_nuc(seq1[linha],seq2[coluna],match,mismatch)
        cima = score_matrix[linha - 1][coluna] + gap
        esquerda = score_matrix[linha][coluna - 1] + gap

        max_score = max(diagonal, esquerda, cima)
        score_matrix[linha][coluna] = max_score

        if max_score == diagonal: traceback_matrix[linha][coluna] = '↖'
        elif max_score == cima: traceback_matrix[linha][coluna] = '↑'
        elif max_score == esquerda: traceback_matrix[linha][coluna] = '←'

  else: # alinhamento de proteinas
    for linha in range(1, nrows):
      for coluna in range(1, ncols):
        diagonal = score_matrix[linha - 1][coluna - 1] + score_subst_prot(matriz,seq1[linha],seq2[coluna])
        cima = score_matrix[linha - 1][coluna] + gap
        esquerda = score_matrix[linha][coluna - 1] + gap

        max_score = max(diagonal, esquerda, cima)
        score_matrix[linha][coluna] = max_score

        if max_score == diagonal: traceback_matrix[linha][coluna] = '↖'
        elif max_score == cima: traceback_matrix[linha][coluna] = '↑'
        elif max_score == esquerda: traceback_matrix[linha][coluna] = '←'

  if matrizes:
      print("Matriz de Pontuação:")
      print_matrix(seq2,seq1,score_matrix)
      print("Matriz de Rastreamento:")
      print_matrix(seq2,seq1,traceback_matrix)
      return score_matrix[-1][-1],reconstrucao(seq1,seq2,traceback_matrix,matrizes)
    
  else:
      return score_matrix[-1][-1],reconstrucao(seq1,seq2,traceback_matrix,matrizes)
```

## Tarefa 6 - Blast

<div align='justify'>

Na tarefa 6, foi propostoa construção do algoritmo Blast. Neste notebook, discutiu-se os algoritmos heurísticos e as ferramentas existentes para esta tarefa e procedeu-se à implementação de uma versão simplificada do algoritmo para comparação de sequências biológicas mais popular, o BLAST.<br>
A tentativa de inferir a função a partir de sequências semelhantes é uma forma de abordar esta tarefa, em que se tentou
identificar sequências que sejam semelhantes à sequência alvo (de consulta) e que já estejam anotadas com uma determinada função.<br> 
O algoritmo BLAST é atualmente o mais utilizado para efetuar pesquisas em bases de dados de sequências de DNA ou de proteínas. O objetivo é procurar bons alinhamentos entre uma sequência de consulta e as sequências de uma base de dados definida. A ideia básica dos algoritmos é utilizar "palavras" curtas (ou seja, sub-sequências) e procurar correspondências de elevada semelhança destas palavras na consulta e nas sequências da base de dados. Estas correspondências formarão uma base que pode ser
que pode ser alargada, em ambas as direcções, para obter alinhamentos de alta qualidade de maior dimensão.<br>

Inicialmente, há uma string (a query), que é a sequência que queremos procurar dentro de uma sequência alvo (seq); e um parâmetro do algoritmo, W (tamanho), que nos indica as sub-sequências de comprimento w que ocorrem na sequência de consulta.<br>
Começou-se por construir a função parse_query, que recebe uma sequência (query) e realiza as operações de remoção de espaços em branco, converção da sequência para maiúsculas e verificação se a sequência contém apenas as letras permitidas (A, C, G, T). Assim, consegue-se que o input dado pelo utilizador seja válida, pois a função retorna a a sequência processada.<br>
De seguida, construiu-se a função query_map, que recebe a sequência que se quer procurar dentro de uma sequência alvo (query) e o tamanho da janela de procura(parâmetro w), e  devolve um dicionário que tem como chaves as palavras da consulta e como valores as listas de posições onde estas ocorrem como valores.<br>
Passou-se para a construção da função hits, que recebe um dicionário que contem um mapeamento de queries para offsets (resultado da função query_map) e uma sequência alvo, e realiza buscas utilizando as queries na sequência alvo, retornando uma lista de tuplos, onde cada tuplo contém dois valores: o offset na query e o offset na sequência.<br>
Para o auxílio desta função, foram ainda construidas duas funções auxiliares: procura, que procura uma subsequência específica (codao) numa sequência dada (seq) e retorna uma lista contendo os offsets onde a subsequência é encontrada; e criar_tuplos, que cria tuplos contendo o offset fornecido e os elementos da lista de índices.<br>
Para estender o hit na sequência até que não seja mais possível, mantendo um número mínimo de matches, montou-se a função extend_hit.
Por último, com a função best_hit, realizou-se extensões de hits, retornando o melhor hit de todos.Esta identifica o melhor alinhamento entre a consulta e uma dada sequência, com todos os resultados de tamanho w. O que tivar com a melhor pontuação global (maior número de correspondências) será selecionado.<br>

```python
def parse_query(query : str) -> str:
  """
    Esta função recebe uma sequência (query) e realiza as seguintes operações:
      1. Remove espaços em branco.
      2. Converte a sequência para maiúsculas.
      3. Verifica se a sequência contém apenas as letras permitidas (A, C, G, T).

    Parameters
    ----------
    query : str
      a sequência que será processada.

    Returns
    -------
    str
      a sequência processada, após a remoção de espaços, conversão para maiúsculas e validação das letras permitidas.

    Raises
    ------
    ValueError
      se a sequência for vazia ou conter caracteres não permitidos.
  """

  query = query.replace(" ", "").upper()

  if query == "": 
      raise ValueError("A sequência não pode ser vazia")
    
  letras_permitidas = set(['A', 'C', 'G', 'T'])

  if set(query) <= letras_permitidas:
      return query

  else:
      raise ValueError("A sequência deve conter apenas as letras A, C, G, T")

def query_map(query : str, w: int) -> dict[str, list[int]]:
  """
  Esta função recebe a sequência que queremos procurar dentro de uma sequência alvo (query) devolve um dicionário que
  tem como chaves as palavras da consulta e como valores as listas de posições onde estas ocorrem como valores.

  Parameters
  ----------
  query : str
    a string que procuramos.

  w : int
    o tamanho da janela. O valor de w por padrão é 3.

  Returns
  -------
  dict : str, list[int]
    Devolve um dicionário:
      cuja chave são todas as strings de tamanho w da query e;
      cujo valor é uma lista com todos os offsets onde aparecem.

  Raises
  ------
  TypeError
    A sequência deve ser uma string.
  
  ValueError
    O tamanho da janela deve ser um inteiro positivo
  """
  if not isinstance(query, str):
    raise TypeError("A sequência deve ser uma string")
  
  if not isinstance(w, int) or w <= 0:
    raise ValueError("O tamanho da janela deve ser um inteiro positivo")

  query = parse_query(query)

  res={}
  for I in range(len(query)):
    sub_string=query[I:I+w]

    if sub_string not in res and len(sub_string)==w:
      res[sub_string]=[I]
    elif sub_string in res and len(sub_string)==w:
      res[sub_string].append(I)
  return res

qm = query_map(query, w)

def procura(codao : str, seq : str) -> list[int]:
  """
  Função auxiliar da função hits.
  Esta função procura uma subsequência específica (codao) numa sequência dada (seq) e
  retorna uma lista contendo os offsets onde a subsequência é encontrada.

  Parameters
  ----------
  codao : str
    a subsequência que estamos a procurar na sequência.

  seq : str
    a sequência na qual estamos a procurar a subsequência.

  Returns
  -------
  list[int]
    uma lista que contem os offsets (índices) onde a subsequência é encontrada na sequência.
  """
  tm = len(codao)
  indices = []
  for i in range(len(seq)):
      sub_string = seq[i: i + tm]
      if codao == sub_string:
          indices.append(i)
  return indices

def criar_tuplos(indices : list[int], offset : int) -> list[tuple[int, int]]: 
  """
  Função auxiliar da função hits.
  Esta função cria tuplos contendo o offset fornecido e os elementos da lista de índices.

  Parameters
  ----------
  indices : list[int]
    lista que contem os índices para os quais os tuplos serão criados.

  offset : int
    o valor do offset que será incluído nos tuplos.

  Returns
  -------
  list[tuple[int, int]]
    Lista de tuplos que contem o offset e cada elemento da lista de índices.
  """
  tuplos = []

  for l in indices:
      tuplos.append((offset, l))
  return tuplos

def hits(qm : dict[str, list[int]], seq : str) -> list[tuple[int, int]]:
  """
  Esta função recebe um dicionário que contem um mapeamento de queries para offsets e uma sequência alvo,
  realiza buscas utilizando as queries na sequência alvo e retorna uma lista de tuplos,
  onde cada tuplo contém dois valores: o offset na query e o offset na sequência.

  Parameters
  ----------
  qm : dict[str, list[int]]
    o que é devolvido ao invocar a função query_map.

  seq : str
    a sequência alvo.

  Returns
  -------
  list : tuple[int, int]
    Devolve uma lista em que cada elemento é um tuplo com dois valores:
      1. O Offset na query;
      2. O offset na seq.

  Raises
  ------
  TypeError
    A sequência deve ser uma string.
  """
  if not isinstance(seq, str):
    raise TypeError("A sequência deve ser uma string")
  hits_list = [] 

  seq = parse_query(seq)

  for query, offsets in qm.items():
      # seq <- indices query
      res = procura(query, seq)
      # [0 , 12, 15] 0 -> 0,0 0,12 0,15
      for offset in offsets:
          hits_list += criar_tuplos(res, offset) 

  return hits_list

res = hits(qm, seq)

def extend_hit(query : str, seq : str, hit : tuple[int, int], w: int) -> tuple[int, int, int, int]:
  """
  Esta função recebe uma query, uma sequência alvo, um hit e um tamanho de janela,
  e estende o hit na sequência até que não seja possível mais, mantendo um número mínimo de matches.
      
  Parameters
  ----------
  query : str
    a string que procuramos.

  seq : str
    a sequência alvo.

  hit : tuple[int, int]
    um dos elementos devolvidos pela invocação da função hits.

  w : int
    o tamanho da janela. O valor de w por padrão é 3.

  Returns
  -------
  tuple : [int, int, int, int]
    Devolve um tuplo com:
        1. O offset inicial na query;
        2. O offset inicial na seq;
        3. O tamanho do resultado;
        4. O nº de matches corretos.

  Raises
  ------
  TypeError
    A sequência deve ser uma string.
  ValueError
    O tamanho da janela deve ser um inteiro positivo.
  """
  if not isinstance(query, str):
    raise TypeError("A sequência deve ser uma string")
  if not isinstance(seq, str):
    raise TypeError("A sequência deve ser uma string")
  if not isinstance(w, int) or w <= 0:
    raise ValueError("O tamanho da janela deve ser um inteiro positivo")
  
  query = parse_query(query)
  seq = parse_query(seq)

  num_matches = w
  tam_extensao = w 
  i_min_query = hit[0]
  i_min_seq = hit[1]
  i_max_query = hit[0] + w - 1
  i_max_seq = hit[1] + w - 1
    
  while num_matches >= tam_extensao / 2 and ((i_min_seq > 0 and i_min_query > 0) or (i_max_seq < len(seq) - 1 and i_max_query < len(query) - 1)):
      # esquerda 
      if i_min_query>0 and i_min_seq>0:
          i_min_query -= 1
          i_min_seq -= 1 
          tam_extensao += 1
          if query[i_min_query] == seq[i_min_seq]:
              num_matches += 1
      # direita 
      if i_max_query < len(query) - 1 and i_max_seq < len(seq) - 1:
          i_max_query += 1
          i_max_seq += 1 
          tam_extensao += 1
          if query[i_max_query] == seq[i_max_seq]:
              num_matches += 1
            
  return (i_min_query, i_min_seq, i_max_query - i_min_query + 1, num_matches)

def best_hit(query : str, seq : str, w: int) -> tuple[int, int, int, int]:
  """
  Esta função recebe uma query, uma sequência alvo, e um tamanho de janela,
  realiza extensões de hits usando a função extend_hit, e retorna o melhor hit encontrado.

  Parameters
  ----------
  query : str
    a string que procuramos.

  seq : str
    a sequência alvo.

  w : int
    o tamanho da janela. O valor de w por padrão é 3.

  Returns
  -------
  tuple : [int, int, int, int]
    Devolve um tuplo com:
      1. O offset inicial na query;
      2. O offset inicial na seq;
      3. O tamanho do resultado;
      4. O nº de matches corretos.

  Raises
  ------
  TypeError
    A sequência deve ser uma string.
  ValueError
    O tamanho da janela deve ser um inteiro positivo
  """
  if not isinstance(query, str):
    raise TypeError("A sequência deve ser uma string")
  if not isinstance(seq, str):
    raise TypeError("A sequência deve ser uma string")
  if not isinstance(w, int) or w <= 0:
    raise ValueError("O tamanho da janela deve ser um inteiro positivo")
  
  query = parse_query(query)
  seq = parse_query(seq)

  qm = query_map(query, w)
  hits_list = hits(qm, seq)
  
  best_hit_result = None
  best_score = 0

  for hit in hits_list:
      extended_hit = extend_hit(query, seq, hit, w)
      score = extended_hit[3]

      if score > best_score or (score == best_score and extended_hit[2] < best_hit_result[2]):
        best_hit_result = extended_hit
        best_score = score

  return best_hit_result
```

    onde cada dicionário contém as pontuações logarítmicas para cada base ou aminoácido em uma posição específica.

```python
def alfabeto(tipo_seq : str) -> list(str):
    '''
    Esta função recebe o tipo de uma sequência e devolve o seu respetivo alfabeto.

    Parameters
    ---------
    tipo_seq : str
        Tipo válido de uma sequência biológica. Pode assumir os valores de DNA, RNA, ou
        proteína.

    Returns
    -------
    list(str)
        Composto pelo tipo de sequência e o seu respetivo alfabeto
    
    Raises
    ------
    ValueError
        O tipo não é válido. Os tipos de cadeia válida são: DNA, RNA ou proteina.
    '''
    from unidecode import unidecode
    if tipo_seq.upper()=="DNA":
        DNA="ACGT"
        return [tipo_seq.upper(),DNA]
    elif tipo_seq.upper()=="RNA":
        RNA="ACGU"
        return [tipo_seq.upper(),RNA]
    elif unidecode(tipo_seq.upper()) in ["PROTEIN","PROTEINA","PROT"]:
        protein="ABCDEFGHIKLMNPQRSTVWYZ"
        return ["Proteína", protein]
    else:
        raise ValueError(f"O tipo {tipo_seq} não é válido. Os tipos de cadeia válida são: DNA, RNA ou proteina") 

def pwm(seqs : list(str), tipo_seq : str, pseudo : int = 0) -> list(dict[str,float]):
    '''
    Esta função recebe uma lista de sequências alinhadas e devolve uma matriz desingada de position weigth matrix (PWM), onde se representa
    a probabilidade de cada letra de um dado alfabeto biológico estar presente numa determinada posição.

    Parameters
    ----------
    seqs : list(str)
        Contêm um conjunto de sequências alinhadas em formato de string.

    tipo_seq : str
        Tipo da sequência que se pretende realiazar a PWM. Pode assumir o valor de DNA, RNA ou proteína.

    pseudo : int ou float
        Valor a utilizar para realizar a pseudocontagem. Por padrão o valor é 0.

    Returns
    -------
    PWM : list(dict)
        Valor da probabilidade de cada letra do alfabeto estar presente em cada posição das sequências.
    
    Raises
    ------
    AssertionError
        Na sequencia {seq}, o elemento {elem} na posição {pos+1} não se encontra no alfabeto do {alfa[0]}.
        A sequência {seq} não possui o mesmo tamanho que as restantes.
    '''
    alfa=alfabeto(tipo_seq)
    
    tamanho=[]
    
    for I,seq in enumerate(seqs):
        tamanho.append(len(seq))
        for pos,elem in enumerate(seq):
            assert elem in alfa[1],f"Na sequencia {seq}, o elemento {elem} na posição {pos+1} não se encontra no alfabeto do {alfa[0]}"
        assert tamanho[I]==tamanho[I-1],f"A sequência {seq} não possui o mesmo tamanho que as restantes"

    elementos_coluna=list(zip(*seqs))

    PWM=[]

    for coluna in elementos_coluna:
        contagem={}
        for elem in alfa[1]:
            contagem[elem]=(coluna.count(elem) + pseudo)/(len(seqs)+len(alfa[1])*pseudo)
        
        PWM.append(contagem)  
    
    return PWM

def prob_seq (seq : str, pwm : list(dict[str,float])) -> float:
    '''
    Esta função calcula a probabilidade de uma determinada sequencia ter sido gerada pela PWM.

    Parameters
    ---------
    seq:str
        Sequência que se pretende calcular a probabilidade de ter sido gerada pela PWM.
    
    pwm:list(dict[str,floar])
        Corresponde às probabilidades de cada elemento aparecer nas várias posições da sequência.

    Returns
    -------
    prod : float
        Probabilidade de a sequencia fornecida ter sido gerada pela PWM.

    Raises
    ------
    AssertionError
        A sequência não é do mesmo tamanho que a PWM.
    '''
    assert len(seq)<=len(pwm),f"A sequência {seq} não é do mesmo tamanho que a PWM"

    prod=1
    for I,elem in enumerate(seq):
        prod=prod*pwm[I][elem]
    return prod

def seq_mais_provavel(seq : str, pwm : list(dict[str,float]), resultado : str = "Max") -> str:
 '''
  Esta função recebe uma sequência e utiliza a PWM para determinar qual a subsequência mais provável de ter sido gerada pela PWM.

  Parameters
  --------
  seq : str
    sequência na qual se pretende analisar a probablidade de cada subsequência ter sido gerada pela PWM.

  pwm : list(dict[str,float])
    PWM de um determinado alinhamento

  resultado : 
  str
    Corresponde ao valor que se pretende obter, por exemplo, quando assume o valor "Máximo" devolve a(s) subsequência(s) 
    com maior probabilidade de ter sido gerada pela PWM, quando assume o valor mínimo devolve a(s) sequência(s) com menor probabildade
    de serem formadas pela PWM

  Returns
  -------
  str
    Corresponde à sequência de maior/menor probabilidade
  Raises
  ------
  AssertionError
    A sequência não pode ser utilizada para determinar a sequência mais próvavel. A sequência deve ter, no mínimo, o mesmo tamnho da PWM.
 '''

 assert len(seq)>=len(pwm),f'''
 A sequência {seq} não pode ser utilizada para determinar a sequência mais próvavel. A sequência deve ter, no mínimo, o mesmo tamnho da PWM
 '''
 w=len(pwm)
 snips=[seq[P:P+w] for P in range(len(seq)-w+1)]
 res={}
 for snip in snips:
    prob=prob_seq(snip,pwm)
    res[snip]=prob
  
 if resultado.upper() in ['MAX','MAXIMO']:
    return max(res, key=res.get)
 
 if resultado.upper() in ['MIN','MINIMO']:
     return min(res, key=res.get)
 
def pssm(seqs : list(str), tipo_seq : str, pseudo : int = 1) -> list(dict):
    '''
    Calcula a Matriz de Pontuação Específica ao Posicionamento (PSSM) para um conjunto de sequências.

    Parameters
    --------
    seqs : list(str)
      Lista de sequências a serem processadas.

    tipo_seq : str
      Tipo de sequência ('DNA', 'RNA' ou 'PROTEIN').

    pseudo : int
      Valor a ser adicionado para evitar problemas com contagens zero

    Returns
    -------
    list of dict
      Lista de dicionários representando a PSSM.
    
    Raises
    ------
    AssetionError
      Na sequencia, o elemento na posição {pos+1} não se encontra no alfabeto do {alfa[0]}.
      A sequência {seq} não possui o mesmo tamanho que as restantes.
    '''
    import math
    alfa=alfabeto(tipo_seq)
    tamanho=[]
    for I,seq in enumerate(seqs):
        tamanho.append(len(seq))
        for pos,elem in enumerate(seq):
            assert elem in alfa[1],f"Na sequencia {seq}, o elemento {elem} na posição {pos+1} não se encontra no alfabeto do {alfa[0]}"
        assert tamanho[I]==tamanho[I-1],f"A sequência {seq} não possui o mesmo tamanho que as restantes"

    elementos=list(zip(*seqs))

    PSSM=[{b:math.log2(((elemento.count(b)+pseudo)/(len(seqs)+len(alfa[1])*pseudo))/(1/len(alfa[1]))) for b in alfa[1]}  for elemento in elementos]

    return PSSM
```

<div align='justify'>

Na tarefa 8, .<br>

find_min_distance - Esta função percorre a matriz e encontra os índices correspondentes à menor distância numérica presente na matriz.

cal_novas_dis - Esta função recebe uma matriz, os índices de duas sequências semelhantes e o nome resultante da combinação dessas
    sequências. Calcula as novas distâncias entre a sequência agregada e as demais sequências presentes na matriz,
    armazenando o resultado em um dicionário aninhado.
nova_matriz - Esta função recebe uma matriz de distâncias, os índices de duas sequências semelhantes e o nome resultante da
    combinação dessas sequências. Ela gera uma nova matriz de distâncias após agregar as sequências semelhantes,
    adicionando a última linha à matriz original.

```python
import re

def leitura_seqs(nome_file : str) -> dict[str,str]:
    """
    Lê sequências a partir de um arquivo em formato FASTA.

    Parameters
    ----------
    nome_file : str
      Nome do arquivo (sem a extensão .fa) contendo as sequências no formato FASTA.

    Returns
    -------
    dict
      Um dicionário onde as chaves são os nomes das sequências e os valores são as sequências correspondentes.
    """
    with open(nome_file+'.fa','r') as file:
        seqs={}
        seq=''
        nome=''
        estado=False
        for line in file:
            if '>' in line:
                if len(nome)!=0 and len(seq)!=0:
                    seq=seq.replace('\n','')
                    seqs[nome]=seq
                    seq=''
                    estado=False

                res=re.findall(r'[A-Z_\d.]+',line)
                nome=res[0]
                estado=True
                continue

            if estado:
                seq+=line

    seq=seq.replace('\n','')
    seqs[nome]=seq
    return seqs

def calculate_distance(seq1 : str, seq2 : str) -> int:
    '''
    Calcula a distância de duas sequências utiliznado o algortimo de Needleman Wunsch

    Parameters
    ----------
    seq1 :str
        Sequência que se pretende calcular a distância
    
    seq2: str
        Sequência com a qual se vai alinhar a sequência 1

    Returns
    ---------
    int
        Corresponde ao resultado da distância que será o simétrico do score obtido no algoritmo de Needleman Wunsch
    '''
    return -needleman_wunsch(seq1, seq2, seq_tipo='protein',matrizes=False)[0]

def matriz_distancias(seqs : dict[str,str]) -> list[list]:
    """
    Calcula a matriz de distâncias entre as sequências fornecidas.

    Parameters
    ----------
    seqs : dict
      Um dicionário onde as chaves são os nomes das sequências e os valores são as próprias sequências.

    Returns
    -------
    list of list
      Uma matriz de distâncias entre as sequências.

    Raises
    ------
    ValueError
      Se o dicionário de sequências estiver vazio.
    """
    nlinhas= len(seqs)+ 1
    ncolunas=len(seqs)+ 1
    matriz=[[0]*ncolunas for linhas in range(nlinhas)] #inicializar a matriz
    n_seqs=['-']

    for seq in seqs: #nome das sequências em estudo
        n_seqs.append(seq)
    
    matriz[0]=[seq for seq in n_seqs] #prencher a primeira linha com os nomes da sequências
    for linha in range(1,nlinhas):
        matriz[linha][0]=n_seqs[linha] #prencher a primeira coluna de cada linha com o nome da sequência

    for linha in range(1,nlinhas):
        for coluna in range(1,ncolunas):
            if linha==coluna:
                pass

            else:
                matriz[linha][coluna]=calculate_distance(seqs[n_seqs[linha]],seqs[n_seqs[coluna]])

    return matriz

def find_min_distance(matrix : list[list]) -> tuple[int,int]:
    """
    Encontra os índices correspondentes à menor distância numa matriz.

    Parameters
    ----------
    matrix : list of list
      Uma matriz contendo distâncias.

    Returns
    -------
    tuple of int
      Um tuplo de dois inteiros representando os índices correspondentes à menor distância.

    Raises
    ------
    ValueError
      Se a matriz estiver vazia ou não contiver valores numéricos.
    """
    min_distance = float('inf')
    min_indices = (0, 0)

    for i,_ in enumerate(matrix):
        for j,_ in enumerate(matrix[i]):
            if type(matrix[i][j])!=int:
                pass
            else:
                if matrix[i][j] < min_distance:
                    min_distance = matrix[i][j]
                    min_indices = (i,j)

    return min_indices

def agru_seq_sem(seq_similar : tuple[int,int], matriz : list[list]) -> str:
    """
    Agrupa duas sequências semelhantes com base nas informações fornecidas por uma matriz.

    Parameters
    ----------
    seq_similar : tuple of int
      Um tuplo que contem os índices das sequências semelhantes na matriz.

    matriz : list of list
      Uma matriz que contem informações sobre as sequências.

    Returns
    -------
    str
      O nome resultante da combinação das sequências semelhantes.

    Raises
    ------
    ValueError
      Se os índices fornecidos estiverem fora dos limites da matriz ou não forem inteiros.
    """
    linha=seq_similar[0]
    coluna=seq_similar[1]
    nome=matriz[linha][0]+' & '+ matriz[0][coluna]
    return nome

def cal_novas_dis(matriz : list[list], seq_similar : tuple[int,int], nome : str) -> dict[str,dict[str,float]]:
    """
    Calcula as novas distâncias após agregar duas sequências semelhantes numa matriz.

    Parameters
    ----------
    matriz : list of list
      Matriz que contem informações sobre as sequências.

    seq_similar : tuple of int
      Um tuplo que contem os índices das sequências semelhantes na matriz.

    nome : str
      O nome resultante da combinação das sequências semelhantes.

    Returns
    -------
    dict of dict
      Um dicionário que contem as novas distâncias calculadas após a agregação das sequências.

    Raises
    ------
    ValueError
      Se os índices fornecidos estiverem fora dos limites da matriz ou não forem inteiros.
    """
    seq1=matriz[seq_similar[0]][0] # nome da sequencia 1 a agregar
    seq2=matriz[0][seq_similar[1]] # nome da sequencia 2 a agregar
    pos_coluna=[] #colunas que correspondem as sequencias a agregar
    calc={}
    calc[nome]={}

    for I in range(len(matriz)):
        if I ==0:
            for J,elem in enumerate(matriz[I]): #selecionar as colunas de interesse
                if elem==seq1 or elem==seq2:
                    pos_coluna.append(J)

        else:
            if matriz[I][0]==seq1 or matriz[I][0]==seq2: #saltar as sequências a agregar
                pass
            else:  # calcula da distância para as sequências agregadas e as restantes
                dist=0
                nome_seq=matriz[I][0]
                for col in pos_coluna:
                    dist+=matriz[I][col]
                dist=dist/2
                calc[nome][nome_seq]=dist
    
    return calc

def nova_matriz(matriz : list[list], seq_similar : tuple[int,int], nome : str) -> list[list]:
    """
    Gera uma nova matriz de distâncias após agregar duas sequências semelhantes.

    Parameters
    ----------
    matriz : list of list
      Matriz original que contem informações sobre as distâncias entre as sequências.

    seq_similar : tuple of int
      Um tuplo que contem os índices das sequências semelhantes na matriz.

    nome : str
      O nome resultante da combinação das sequências semelhantes.

    Returns
    -------
    list of list
      Uma nova matriz de distâncias após a agregação das sequências.

    Raises
    ------
    ValueError
      Se os índices fornecidos estiverem fora dos limites da matriz ou não forem inteiros.
    """
    seq1=matriz[seq_similar[0]][0] # nome da sequencia a agregar
    seq2=matriz[0][seq_similar[1]] # nome da sequencia a agregar
    novas_dist=cal_novas_dis(matriz,seq_similar,nome)
    nova_matriz=[] #iniciar a nova matriz
    pos_col=[] #colunas das sequências a agregar

    for I in range(len(matriz)): #copia da matriz de distâncias original
        linha=[]
        if I==0:
            for J,elem in enumerate(matriz[I]):
                if elem==seq1 or elem==seq2:
                    pos_col.append(J)
                    pass
                else:
                    linha.append(elem)
            nova_matriz.append(linha)
            continue
        
        else:
            if matriz[I][0]==seq1 or matriz[I][0]==seq2: # as linhas que correspondem as seqs a agregar são ignoradas
                continue

            else: 
                for J,elem in enumerate(matriz[I]): #seleção das colunas de interesse para um certa linha
                    if J not in pos_col: #as colunas que correspondem as seqs a agregar são ignoradas
                        linha.append(elem)

        nova_matriz.append(linha)

    nova_matriz[0]=nova_matriz[0] + [nome] #adiconar na primeira linha as seqs agregadas

    linhas=len(nova_matriz)
    for I in range(1,linhas): #adicionar a partir da linha 1 o valor da nova distancia
        for seq in novas_dist[nome]:
            if seq== nova_matriz[I][0]:
                nova_matriz[I].append(novas_dist[nome][seq])
    
    ult_linha=[] #construção da ultima linha
    ult_linha.append(nome)
    for I in range(1,linhas):
        ult_linha.append(nova_matriz[I][-1])
    ult_linha.append(0)

    nova_matriz.append(ult_linha) #adicicao da ultima linha à matriz

    return nova_matriz

def arvore(nome_file : str) -> list:
    """
    Constrói uma árvore filogenética iterativa com base num arquivo de sequências.

    Parameters
    ----------
    nome_file : str
      O nome do arquivo que contem as sequências a serem processadas.

    Returns
    -------
    list
      Uma lista que representa a árvore filogenética iterativa.

    Raises
    ------
    ValueError
      Se ocorrerem problemas ao processar as sequências ou calcular distâncias.
    """
    res=[]
    seqs=leitura_seqs(nome_file)
    mat_dist=matriz_distancias(seqs)
    min_dist=find_min_distance(mat_dist)
    nome=agru_seq_sem(min_dist,mat_dist)
    print('Cluster 1: {}'.format(nome))
    res.append(nome)
    cont=1

    while len(res)!=len(seqs)-1:
        cont+=1
        mat_dist=nova_matriz(mat_dist,min_dist,nome)
        min_dist=find_min_distance(mat_dist)
        nome=agru_seq_sem(min_dist,mat_dist)
        res.append(nome)
        print('Cluster {}: {}'.format(cont,nome))
    
    return res
```
